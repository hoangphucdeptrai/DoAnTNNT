K-Mean là gì?

K-Means Clustering là một trong những thuật toán nổi tiếng. Phương pháp này đơn
giản nhưng đặc biệt hiệu quả trong bài toán học không giám sát, khi mà dữ liệu của bạn
chưa được phân loại. Mục tiêu của thuật toán này là chia nhỏ data của bạn thành k groups
dựa trên đặc trưng được cung cấp. Các điểm dữ liệu được xếp vào trong từng nhóm dựa trên
sự giống nhau về đặc điểm nhận dạng.

Ứng dụng của k-mean:
1. Phân khúc khách hàng (Customer Segmentation) - Phổ biến nhấtTrong kinh doanh và Marketing, việc hiểu khách hàng là yếu tố sống còn. K-Means giúp chia danh sách khách hàng khổng lồ thành các nhóm nhỏ có đặc điểm giống nhau.Cách hoạt động: Dựa trên dữ liệu như: lịch sử mua hàng, độ tuổi, tổng số tiền đã chi tiêu, tần suất ghé thăm website.Kết quả:Nhóm VIP: Chi tiêu nhiều, cần chăm sóc đặc biệt.Nhóm tiết kiệm: Chỉ mua khi giảm giá -> Gửi coupon khuyến mãi.Nhóm rời bỏ: Lâu không quay lại -> Cần email lôi kéo.Lợi ích: Tối ưu hóa chiến dịch Marketing (Targeted Marketing) thay vì gửi tin nhắn rác cho tất cả mọi người.
2. Xử lý ảnh và Thị giác máy tính (Image Processing)K-Means cực kỳ mạnh mẽ trong việc xử lý dữ liệu điểm ảnh (pixel).Nén màu ảnh (Color Quantization):Một bức ảnh có thể có hàng triệu màu. K-Means giúp giảm số lượng màu xuống còn $k$ màu (ví dụ: 64 màu) đại diện nhất.Mỗi pixel sẽ được thay thế bằng màu của tâm cụm (centroid) gần nhất. Điều này giúp giảm đáng kể dung lượng ảnh mà vẫn giữ được nội dung chính.Phân vùng ảnh (Image Segmentation):Tách đối tượng ra khỏi nền (background). Ví dụ: Trong ảnh chụp X-quang, dùng K-Means để tách vùng khối u (có độ sáng khác biệt) ra khỏi các mô xung quanh.
3. Phát hiện bất thường (Anomaly/Fraud Detection)Trong an ninh mạng và tài chính, K-Means được dùng để phát hiện các hành vi gian lận hoặc tấn công.Cách hoạt động:Gom nhóm các hành vi giao dịch/truy cập "bình thường".Bất kỳ điểm dữ liệu nào nằm quá xa các tâm cụm (centroids) hoặc không thuộc về cụm nào một cách rõ ràng sẽ bị đánh dấu là "bất thường" (Outlier).Ứng dụng: Phát hiện giao dịch thẻ tín dụng giả mạo, phát hiện xâm nhập mạng (Intrusion Detection System).
4. Phân loại văn bản và Tin tức (Document Clustering)K-Means giúp tổ chức lượng lớn dữ liệu văn bản không có cấu trúc.Cách hoạt động: Biến đổi văn bản thành các vector số (dùng TF-IDF hoặc Word Embeddings). K-Means sẽ gom các vector này lại.Ứng dụng:Google News: Tự động gom các bài báo từ nhiều nguồn khác nhau nói về cùng một sự kiện (ví dụ: "Bão số 3") vào một nhóm.Lọc thư rác (Spam filtering): Gom nhóm các email có nội dung giống spam.
5. Gợi ý kích cỡ quần áo (Size Recommendation)Các hãng thời trang thu thập dữ liệu chiều cao, cân nặng, số đo các vòng của hàng nghìn người.Dùng K-Means gom nhóm lại để xác định ra các kích cỡ chuẩn S, M, L, XL phù hợp với đại đa số khách hàng mục tiêu của họ.
6. Tối ưu hóa Logistics và Vận tảiBài toán: Bạn có 100 đơn hàng cần giao và 5 tài xế.Ứng dụng: Dùng K-Means (với $k=5$) để gom các địa điểm giao hàng thành 5 khu vực địa lý tập trung nhất. Mỗi tài xế sẽ phụ trách một cụm, giúp tối ưu quãng đường di chuyển.

K-NN là gì?

Thuật toán K láng giềng gần nhất là một kĩ thuật học có giám sát
(supervised learning) dùng để phân loại quan sát mới bằng cách tìm điểm tương
đồng giữa quan sát mới này với dữ liệu sẵn có.

Ứng dụng của K-NN:

1. Hệ thống gợi ý (Recommender Systems) - Ứng dụng nổi tiếng nhấtCác nền tảng như Netflix, Amazon, Spotify sử dụng nguyên lý tương tự K-NN để gợi ý sản phẩm.
Cơ chế (Collaborative Filtering):Nếu người dùng A thích phim $X, Y$.Người dùng B cũng thích phim $X, Y$ và thích thêm phim $Z$.Khoảng cách giữa A và B rất nhỏ (họ là "láng giềng").=> Hệ thống sẽ gợi ý phim $Z$ cho người dùng A.Ưu điểm: Không cần hiểu nội dung phim, chỉ cần dựa vào hành vi tương đồng của người dùng.

2. Nhận dạng mẫu & Thị giác máy tính (Pattern Recognition)Trước khi Deep Learning (CNN) thống trị, K-NN là công cụ mạnh mẽ trong xử lý ảnh đơn giản.Nhận dạng chữ viết tay (OCR):Dùng để đọc mã bưu điện trên phong bì hoặc số tiền trên séc ngân hàng.Một chữ số viết tay mới sẽ được so sánh từng pixel với cơ sở dữ liệu các chữ số mẫu. Nếu nó "gần" với các hình mẫu số 8 nhất, nó được đoán là số 8.Nhận diện khuôn mặt (Cơ bản): So sánh các đặc trưng khuôn mặt (khoảng cách mắt, mũi, miệng) với database để tìm người giống nhất.
3. Tài chính và Ngân hàngNgành ngân hàng cần sự minh bạch trong quyết định, và K-NN rất dễ giải thích.Đánh giá rủi ro tín dụng (Credit Scoring):Khi có hồ sơ vay mới, ngân hàng so sánh với các hồ sơ cũ.Nếu hồ sơ này "nằm gần" những người từng bùng nợ (nhóm nợ xấu) -> Từ chối.Nếu nằm gần nhóm trả nợ đúng hạn -> Duyệt.Phát hiện gian lận (Fraud Detection): Các giao dịch thẻ tín dụng có hành vi khác thường (ví dụ: mua sắm ở nước lạ, số tiền lớn đột biến) sẽ có khoảng cách rất xa so với các giao dịch bình thường của chủ thẻ đó -> Cảnh báo.
4. Y tế và Chẩn đoán bệnhChẩn đoán dựa trên triệu chứng:Dữ liệu bệnh nhân (nhịp tim, huyết áp, chỉ số máu, tuổi...) được xem là một điểm dữ liệu.K-NN tìm $k$ bệnh nhân cũ có chỉ số tương tự. Nếu đa số họ bị bệnh tiểu đường, bệnh nhân mới cũng có nguy cơ cao bị tiểu đường.Phân loại biểu hiện gen: Dùng trong tin sinh học để phân loại các loại ung thư dựa trên dữ liệu di truyền.
5. Xử lý dữ liệu (Data Preprocessing)Đây là một ứng dụng kỹ thuật cực kỳ hữu ích cho Developer/Data Scientist: Điền giá trị khuyết (Imputation).Vấn đề: Dữ liệu thu thập thường bị thủng (missing values, NaN).Giải pháp (KNN Imputer): Thay vì điền bằng số trung bình (mean) của toàn bộ cột (kém chính xác), ta tìm $k$ dòng dữ liệu giống dòng bị thiếu nhất, rồi lấy trung bình giá trị của $k$ dòng đó để điền vào chỗ trống.Ví dụ: Một người bị thiếu thông tin "Tuổi". Ta tìm những người có thu nhập, nghề nghiệp, giới tính giống người đó nhất, rồi lấy tuổi trung bình của họ điền vào.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Tạo dữ liệu giả lập (3 nhóm rõ rệt)
# n_samples=300: 300 điểm dữ liệu
# centers=3: 3 tâm cụm gốc
# cluster_std=0.60: độ phân tán của cụm
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# 2. Cấu hình và huấn luyện mô hình K-Means
# init='k-means++': thuật toán chọn tâm khởi tạo thông minh giúp hội tụ nhanh hơn
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, random_state=42)
kmeans.fit(X)

# 3. Lấy kết quả
y_kmeans = kmeans.predict(X)       # Nhãn của từng điểm (0, 1, hoặc 2)
centers = kmeans.cluster_centers_  # Tọa độ của 3 tâm cụm tìm được

# 4. Vẽ biểu đồ (Visualization)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis') # Vẽ các điểm dữ liệu
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7, marker='X') # Vẽ tâm cụm (dấu X đỏ)
plt.title("Kết quả phân cụm K-Means (Scikit-Learn)")
plt.show()
```

```python
import numpy as np
from collections import Counter

class KNNClassifier:
    def __init__(self, k=3, weights='uniform'):
        """
        k: số lượng láng giềng
        weights: 'uniform' (tất cả như nhau) hoặc 'distance' (nghịch đảo khoảng cách)
        """
        self.k = k
        self.weights = weights
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        # K-NN là Lazy Learning, chỉ cần lưu dữ liệu lại
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def predict(self, X_test):
        X_test = np.array(X_test)
        predictions = [self._predict_one(x) for x in X_test]
        return np.array(predictions)

    def _predict_one(self, x):
        # 1. Tính khoảng cách từ x đến tất cả các điểm trong tập train
        # Sử dụng khoảng cách Euclid: sqrt(sum((x1-x2)^2))
        distances = np.sqrt(np.sum((self.X_train - x)**2, axis=1))

        # 2. Lấy index của k điểm gần nhất (k-nearest neighbors)
        # argsort trả về index của mảng sau khi sắp xếp tăng dần
        k_indices = np.argsort(distances)[:self.k]

        # Lấy nhãn (y) và khoảng cách (d) của k láng giềng này
        k_nearest_labels = self.y_train[k_indices]
        k_nearest_distances = distances[k_indices]

        # 3. Bầu chọn (Voting)
        if self.weights == 'uniform':
            # CÁCH 1: Uniform - Đếm số lượng nhãn xuất hiện nhiều nhất
            most_common = Counter(k_nearest_labels).most_common(1)
            return most_common[0][0]

        elif self.weights == 'distance':
            # CÁCH 2: Distance Weighted - Tính tổng trọng số cho từng class
            class_weights = {}

            for label, dist in zip(k_nearest_labels, k_nearest_distances):
                # Tránh chia cho 0: nếu khoảng cách cực nhỏ (~0), điểm đó chính là class đó
                if dist <= 1e-5:
                    return label

                weight = 1 / dist # Công thức trọng số nghịch đảo

                if label in class_weights:
                    class_weights[label] += weight
                else:
                    class_weights[label] = weight

            # Trả về nhãn có tổng trọng số lớn nhất
            return max(class_weights, key=class_weights.get)

        else:
            raise ValueError("weights must be 'uniform' or 'distance'")
  # --- CHẠY THỬ NGHIỆM ---

# Tạo dữ liệu giả lập
# Class 0: Nhóm 2 điểm rất gần nhau tại (1, 1) và (1.1, 1.1)
# Class 1: Nhóm 3 điểm ở xa hơn một chút tại (2, 2), (2.1, 2), (2, 2.1)
X_train = [[1, 1], [1.1, 1.1], [2, 2], [2.1, 2], [2, 2.1]]
y_train = [0, 0, 1, 1, 1]

# Điểm cần dự đoán: (1.05, 1.05) -> Rất gần Class 0, nhưng nếu xét k=5 thì Class 1 đông hơn
x_test = [[1.05, 1.05]]

print(f"Dữ liệu train: {len(X_train)} điểm.")
print(f"Điểm cần dự đoán: {x_test[0]}")
print("-" * 30)

# Trường hợp 1: Uniform Weight (k=5)
# Lấy 5 láng giềng -> Có 2 điểm Class 0 và 3 điểm Class 1 -> Class 1 thắng (do đông hơn)
knn_uniform = KNNClassifier(k=5, weights='uniform')
knn_uniform.fit(X_train, y_train)
pred_uniform = knn_uniform.predict(x_test)
print(f"Kết quả Uniform (k=5): Class {pred_uniform[0]} (Sai bản chất, đúng theo đa số)")

# Trường hợp 2: Distance Weight (k=5)
# Lấy 5 láng giềng. Dù Class 1 có 3 phiếu, nhưng 2 phiếu của Class 0 cực gần -> Trọng số cao -> Class 0 thắng
knn_distance = KNNClassifier(k=5, weights='distance')
knn_distance.fit(X_train, y_train)
pred_distance = knn_distance.predict(x_test)
print(f"Kết quả Distance (k=5): Class {pred_distance[0]} (Đúng bản chất do ở gần)")
```

```python
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_classification
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# Cấu hình trang
st.set_page_config(page_title="Demo K-Means & K-NN", layout="wide")

st.title("🔍 Demo Trực Quan: K-Means & K-NN")
st.markdown("""
Ứng dụng này giúp bạn hình dung cách hoạt động của hai thuật toán máy học phổ biến.
Hãy chọn thuật toán ở thanh bên trái để bắt đầu.
""")

# --- Sidebar: Chọn thuật toán ---
algo_option = st.sidebar.selectbox(
    "Chọn thuật toán",
    ("K-Means (Phân cụm)", "K-NN (Phân loại)")
)

# ==========================================
# PHẦN 1: K-MEANS CLUSTERING
# ==========================================
if algo_option == "K-Means (Phân cụm)":
    st.header("1. Thuật toán K-Means")
    st.markdown("Mục tiêu: Gom nhóm dữ liệu không nhãn thành **K** cụm.")

    # --- Cấu hình tham số ---
    col1, col2 = st.columns([1, 3])

    with col1:
        st.subheader("Tham số")
        n_samples = st.slider("Số lượng điểm dữ liệu", 100, 1000, 300)
        n_features = 2 # Giữ cố định 2D để vẽ đồ thị
        centers_true = st.slider("Số cụm thực tế (Dữ liệu gốc)", 2, 8, 4)
        cluster_std = st.slider("Độ phân tán (Nhiễu)", 0.1, 2.0, 0.8)

        st.markdown("---")
        st.subheader("Cấu hình Model")
        k_clusters = st.slider("Chọn K (Số cụm muốn tìm)", 2, 10, 4)

    with col2:
        # 1. Tạo dữ liệu giả lập
        X, y_true = make_blobs(n_samples=n_samples, centers=centers_true,
                               cluster_std=cluster_std, n_features=n_features, random_state=42)

        # 2. Chạy K-Means
        kmeans = KMeans(n_clusters=k_clusters, init='k-means++', n_init=10, random_state=42)
        y_kmeans = kmeans.fit_predict(X)
        centroids = kmeans.cluster_centers_

        # 3. Vẽ biểu đồ
        fig, ax = plt.subplots(figsize=(10, 6))

        # Vẽ các điểm dữ liệu theo màu cụm dự đoán
        scatter = ax.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', alpha=0.6, label='Dữ liệu')

        # Vẽ tâm cụm (Centroids)
        ax.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Tâm cụm (Centroids)')

        ax.set_title(f"Kết quả phân cụm với K={k_clusters}")
        ax.legend()
        st.pyplot(fig)

        st.info(f"💡 **Nhận xét:** Thử thay đổi 'Chọn K' khác với 'Số cụm thực tế' để xem thuật toán cố gắng gom nhóm như thế nào.")

# ==========================================
# PHẦN 2: K-NEAREST NEIGHBORS (K-NN)
# ==========================================
elif algo_option == "K-NN (Phân loại)":
    st.header("2. Thuật toán K-NN")
    st.markdown("Mục tiêu: Phân loại dữ liệu mới dựa trên **K** láng giềng gần nhất.")

    # --- Cấu hình tham số ---
    col1, col2 = st.columns([1, 3])

    with col1:
        st.subheader("Tham số Dữ liệu")
        n_samples_knn = st.slider("Số lượng điểm", 50, 500, 200)
        noise_knn = st.slider("Độ nhiễu (Noise)", 0.0, 1.0, 0.2)

        st.markdown("---")
        st.subheader("Cấu hình Model")
        n_neighbors = st.slider("Số láng giềng (K)", 1, 50, 5)
        weights = st.selectbox("Cách tính trọng số (Weights)",
                               ("uniform", "distance"),
                               help="'uniform': Tất cả láng giềng ngang nhau.\n'distance': Láng giềng gần có tiếng nói lớn hơn.")

    with col2:
        # 1. Tạo dữ liệu giả lập (Classification)
        X, y = make_classification(n_samples=n_samples_knn, n_features=2, n_redundant=0,
                                   n_informative=2, n_clusters_per_class=1, n_classes=3,
                                   random_state=42, class_sep=1.5 - noise_knn)

        # 2. Huấn luyện K-NN
        clf = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
        clf.fit(X, y)

        # 3. Vẽ Ranh giới quyết định (Decision Boundary)
        # Tạo lưới điểm (mesh grid) để dự đoán toàn bộ không gian nền
        h = .05  # độ mịn của lưới
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))

        # Dự đoán màu cho từng điểm trên lưới nền
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        # Vẽ biểu đồ
        fig, ax = plt.subplots(figsize=(10, 6))

        # Vẽ nền màu (Vùng quyết định)
        cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
        cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

        ax.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto', alpha=0.3)

        # Vẽ các điểm dữ liệu thực tế
        scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=50)

        ax.set_title(f"Vùng quyết định K-NN (K={n_neighbors}, Weights='{weights}')")
        st.pyplot(fig)

        # Giải thích trực quan
        st.success(f"""
        **Quan sát:**
        * **Vùng màu nền:** Là khu vực mà thuật toán sẽ dự đoán cho bất kỳ điểm mới nào rơi vào đó.
        * **Khi K nhỏ (ví dụ K=1):** Ranh giới rất gắt, có thể bao quanh từng điểm nhiễu (Overfitting).
        * **Khi K lớn:** Ranh giới mượt mà hơn, tổng quát hơn.
        * **Chế độ '{weights}':** {'Ranh giới mượt hơn tại các vùng giao thoa.' if weights == 'uniform' else 'Ranh giới rất chi tiết, có thể tạo thành các đảo nhỏ bao quanh các điểm nằm sai vùng.'}
        """)
```
